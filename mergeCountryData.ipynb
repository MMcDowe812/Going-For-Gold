```python
import pandas as pd
from pathlib import Path
```

```python
#Get CSV Data
Country_csv = Path("Model data/Country Data.csv")
population_csv = Path("Model data/World Population by country 2024.csv")
world_regions_csv = Path("Model data/continents-according-to-our-world-in-data.csv")

Country_df = pd.read_csv(Country_csv)
population_df = pd.read_csv(population_csv)
world_regions_df = pd.read_csv(world_regions_csv)

Country_df.head()
```

```python
# Drop unused columns
population_df = population_df.drop(columns=["Area (km2)", "Density (/km2)", "Growth Rate", "World %", "World Rank"])

population_df.head()
```

```python
# Drop unused columns
world_regions_df = world_regions_df.drop(columns=["Year"])

world_regions_df.head()
```

```python
# merge dataframes
merged_df = Country_df.merge(population_df, left_on='Country', right_on='Country', how='left')
merged_df = merged_df.merge(world_regions_df, left_on='Country Code', right_on='Code', how='left')
merged_df.head()
```

```python
exportPath = Path('Model data/country_data_merged.csv')
merged_df.to_csv(exportPath)
```
