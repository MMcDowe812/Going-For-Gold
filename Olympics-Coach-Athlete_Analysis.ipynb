```python
#IMPORTING LIBRARIES

import pandas as pd
import numpy as np
```

```python
#CREATING FILE PATHS

ath_file_path = "athletes.csv"
coaches_file_path = "coaches.csv"
medals_file_path = "medals_total.csv"
medallists_file_path = "medallists.csv"
```

```python
#IMPORTING ORIGINAL CSV DATA

raw_coaches = pd.read_csv(coaches_file_path)
raw_athlets = pd.read_csv(ath_file_path)
raw_medals = pd.read_csv(medals_file_path)
raw_medallists = pd.read_csv(medallists_file_path)
```

```python
#viewing raw coaches data

raw_coaches_df = pd.DataFrame(raw_coaches)
#raw_coaches.head(10)
raw_coaches.dtypes

#raw_coaches.dropna()
```

```python
raw_coaches.columns
raw_coaches['name'] = raw_coaches['name'].astype('string')
raw_coaches.dropna()
```

```python
#CREATING A COPY, NEW DATAFRAME AND CSV FOR FURTHER ANALYSIS

raw_coaches_df.copy()

coaches_filter = raw_coaches.drop(columns=['category'])
coach_info_df = pd.DataFrame(coaches_filter)

coach_info_df.to_csv('coach_info.csv', index=False, encoding='utf8')
```

```python
#RAW DATA VIEWING

raw_athlets_df = pd.DataFrame(raw_athlets)
#raw_athlets_df.head(10)
raw_athlets.dropna()
```

```python
#CREATING A COPY, NEW DATAFRAME AND CSV FOR FURTHER ANALYSIS 

raw_athlets_df.copy()
raw_athlets_df.dropna()
raw_athlets_df['name'] = raw_athlets_df['name'].astype('string')
raw_athlets_df['coach'] = raw_athlets_df['coach'].astype('string')
raw_athlets_df['birth_date'] = pd.to_datetime(raw_athlets_df['birth_date'])
athletes_filter = raw_athlets_df[["code", "name", "name_tv", "country_code", "country", "country_full", "height", "weight", "disciplines", "events","birth_date", "coach"]]
athletes_filter.dropna()
athletes_df = pd.DataFrame(athletes_filter)

athletes_df.head(10)

#athletes_df.to_csv('athletes_data.csv', index=False, encoding='utf8')
```

```python
#RAW DATA VIEW AND SAVING TO CSV

raw_medals_df = pd.DataFrame(raw_medals)
raw_medals_df.head(10)

raw_medals_df.to_csv('medal_count.csv', index=False, encoding='utf8')
```

```python
#VIEWING RAW DATA

raw_medallists_df = pd.DataFrame(raw_medallists)
raw_medallists_df.head(10)
```

```python
raw_medallists_df.dropna()
#raw_medals_df.to_csv('medal_count.csv', index=False, encoding='utf8')
```

```python
#CREATING A COPY, NEW DATAFRAME AND CSV FOR FURTHER ANALYSIS 

medallists = raw_medallists.copy()
medallists['name'] = medallists['name'].astype('string')
medallists['birth_date'] = pd.to_datetime(medallists['birth_date'])
medallists_filter = medallists[["medal_type", "medal_code", "name", "country", "country_code", "nationality", "discipline", "event", "birth_date", "code"]]
medallists_df = pd.DataFrame(medallists_filter)
medallists_df.head(10)

medallists_df.to_csv('medallist_info.csv', index=False, encoding='utf8')
```
