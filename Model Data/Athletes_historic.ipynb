```python
# dependencies
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Make a reference to the csv file
Athlete_csv = Path("Olympic_Athlete_bio_history.csv")
medals_csv = Path('Olympic_Athlete_Event_results.csv')

# Import the csv file as a DataFrame
Athlete_df = pd.read_csv(Athlete_csv, encoding="utf-8")
medals_df = pd.read_csv(medals_csv, encoding = "utf-8")

# Check the special characters imported correctly
Athlete_df.head()
```

```python
Athlete_df.columns
```

```python
# Dropping any height,weight that does not have a value.
Athlete_df.dropna()
```

```python
# Make a reference to the csv file
medals_csv = Path('Olympic_Athlete_Event_results.csv')

# Import the csv file as a DataFrame
medals_df = pd.read_csv(medals_csv, encoding = "utf-8")

# Check the special characters imported correctly
medals_df.head()
```

```python
medals_df.columns
```

```python
# looking at Dtypes
medals_df.info()
```

```python
# Casting medal to string
medals_df['medal']=medals_df['medal'].astype('string')
```

```python
# verifying change to medal column
medals_df.info()
```

```python
# replacing NaN with 0
medals_df['medal'] = medals_df['medal'].fillna(0)
medals_df.sample(20)
```

```python
# Replacing the Gold, Silver, Bronze with the value of 1 to indicate medal was won, 0 represents no medal
medals_df['medal'] = medals_df['medal'].replace(to_replace=['Gold', 'Silver', 'Bronze'], value = "1")
```

```python
medals_df.sample(20)
```

```python
# Mergine athlete and even results spreadsheet.
olympic_medal_df = pd.merge(Athlete_df,  
                      medals_df,  
                      on ='athlete_id', 
                      how ='left')
```

```python
# dropping repeat or unnecessary columns
olympic_medal_df = olympic_medal_df.drop(["country_noc_x","country_noc_y","athlete", "result_id" ], axis=1)
olympic_medal_df
```
